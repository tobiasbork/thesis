```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
import pandas as pd

# setup data
seed = np.random.seed(80)
n = 30
y = np.array([-7, 7])
x = np.array([-10, 10])
means = [x.mean(), y.mean()]  
stds = [x.std() / 7, y.std() / 5]
corr = 0.92
covs = [[stds[0]**2          , stds[0]*stds[1]*corr], 
        [stds[0]*stds[1]*corr,           stds[1]**2]] 
X = np.random.multivariate_normal(means, covs, n)

# setup pca
pca = PCA()
pca.fit(X)
transformed_data = pca.transform(X)
pca_df = pd.DataFrame(transformed_data, columns=['PC1','PC2'])
V = pca.components_
proj_operator = np.array([[V[0,0]**2, V[0,0]*V[0,1]],
                 [V[0,1]*V[0,0], V[0,1]**2]])
proj_data = proj_operator.dot(X.T).T
proj_df = pd.DataFrame(proj_data, columns = ['PROJ_1', 'PROJ_2'])


# plot projection lines
plt.figure(figsize = (15,8))
plt.axis('equal')
for i in range(0,n):
    plt.plot([proj_df.loc[i,'PROJ_1'], X[i,0]], [proj_df.loc[i,'PROJ_2'], X[i,1]],'k--', linewidth = 0.75)

# plot principal axis
lambda_1 = pca.explained_variance_[0]*1.5
plt.plot([-lambda_1*pca.components_[0,0],lambda_1*pca.components_[0,0]],[-lambda_1*pca.components_[0,1],lambda_1*pca.components_[0,1]],'k')

    
# plot data points
plt.plot(X[:,0],X[:,1],'k.', markersize = 10)


# plot projected data
plt.plot(proj_df['PROJ_1'],proj_df['PROJ_2'],'r.', markersize = 10)

```
